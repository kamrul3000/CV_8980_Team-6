```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import matplotlib

get_ipython().magic(u'matplotlib inline')
matplotlib.rcParams["savefig.dpi"] = 400 # to get high resolution
```

```python
img = cv2.imread('C:/Future GSU/2018/Computer Vision/Project/Segmentation Technique/brain.png',1)
edges = cv2.Canny(img,150,200)

# Plotting
plt.subplot(121),plt.imshow(img,cmap = 'gray')
plt.title('Original Image'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(edges,cmap = 'gray')
plt.title('Edge Image'), plt.xticks([]), plt.yticks([])
#plt.savefig(C:/Future GSU/2018/Computer Vision/Project/Segmentation Technique/Original_brain.png)
#plt.savefig('./images/Original_Canny.png')
```
